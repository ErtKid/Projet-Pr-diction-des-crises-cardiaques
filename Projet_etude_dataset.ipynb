```python
import matplotlib.pyplot as plt
df = pd.read_csv('heart_disease_health_indicators.csv')

df.describe()
```

```python
ax = sns.histplot(df['Age'], kde=False, color='red')
ax.set(xlabel='Age', ylabel='Nombres de personnes')
plt.show()
```

Aperçu de l'âge des personnes présentes dans cet ensemble de données.

```python
catcol = ['HeartDiseaseorAttack']
plt.figure(figsize=(15,40))
for i,column in enumerate(catcol):
    plt.subplot(len(catcol), 2, i+1)
    plt.pie(x=df[column].value_counts(), labels=df[column].unique(), autopct='%.0f%%')
    plt.title(f"{column}")
    plt.tight_layout()
```

Pourcentage de personnes ayant un problème cardiaque<br>
ici 9% soit environ 22 831 personnes sur 253 880

```python
k = 22 #number of variables for heatmap
cols = df.corr().nlargest(k, 'Smoker')['Smoker'].index
cm = df[cols].corr()
plt.figure(figsize=(20,12))
sns.heatmap(cm, annot=True, cmap = 'RdYlGn')
```
